```python
import spacy
nlp= spacy.load("en_core_web_sm")
```

```python
text= '''"Article 370" most commonly refers to a now-abrogated article of the Indian Constitution that granted special autonomous status to Jammu and Kashmir. Alternatively, it can refer to a 2024 Indian political thriller film about the abrogation of this article. The original Article 370 gave Jammu and Kashmir its own constitution and flag, with the central government's powers limited to defense, foreign affairs, and communication. This special status was revoked by the Indian government in August 2019.
Article 370 (1949-2019)
Purpose: To grant special status to the state of Jammu and Kashmir (J&K).
Special Status: J&K had its own constitution and flag, and was largely autonomous in internal matters except for defense, foreign affairs, and communications.
Residency Rights: Article 35A, a provision derived from Article 370, granted unique rights and privileges to permanent residents of the state, such as property rights.
Abrogation: In August 2019, the Indian government used a presidential order to revoke Article 370.
Legal Status: The Supreme Court upheld the abrogation in December 2023, stating that Article 370 was always intended as a temporary provision and that the abrogation was constitutional.
Article 370 (2024 film)
Genre: Indian Hindi-language political action thriller.
Director: Aditya Suhas Jambhale.
Plot: The film is about the events leading up to the abrogation of Article 370 of the Indian Constitution.
Cast: Stars Yami Gautam Dhar and Priyamani.
Release: It was released on February 23, 2024'''
```

```python
##text to docs conversion
docs=nlp(text)
print(docs)
```

```python
for ent in docs.ents:
  print(ent.text, ": ", ent.label_)
```

```python
##  pos tagging
for token in docs :
  print(f"{token.text}-> {token.pos_}")
```
